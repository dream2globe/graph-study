```python
high_corr = 0.90
mid_corr = 0.70
low_corr = 0.20
ii =  [0.1, 0.5, 0.1, 0.5]
vv = [1, 1, 5, 5]
xrange = 100
```

```python
import numpy as np
from matplotlib import pyplot as plt 
from matplotlib import rc

fig, axes = plt.subplots(2, 2, figsize=(20, 10))
left = 0
right = 100
axes = axes.flatten()

for i, v, ax in zip(ii, vv, axes):
    h_penalty = [i ** ((high_corr)**(n/v)) for n in range(xrange)]
    m_penalty = [i ** ((mid_corr)**(n/v)) for n in range(xrange)]
    l_penalty = [i ** ((low_corr)**(n/v)) for n in range(xrange)]
    ax.set_title(f"Penalty Trend: i={i}, v={v}")
    ax.plot(range(100), h_penalty, 'r-' , label=f'corr: {high_corr}')
    ax.plot(range(100), m_penalty, 'g-' , label=f'corr: {mid_corr}')
    ax.plot(range(100), l_penalty, 'b-' , label=f'corr: {low_corr}')
    ax.grid(color='black', linestyle='-.', linewidth=1)
    ax.set_ylim(0, 1)
    ax.legend()
```
